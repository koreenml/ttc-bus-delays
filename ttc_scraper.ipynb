```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
wikiurl="https://cptdb.ca/wiki/index.php/Toronto_Transit_Commission#Current_fleet_roster"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
ttc_table=soup.find_all('table',{'class':"sortable"})
ttc_table

bus_table = ttc_table[2]
bus_table
```

```python
df=pd.read_html(str(bus_table))
# convert list to dataframe
df=pd.DataFrame(df[0])
df.head()
```

```python
# Clean the table
df.drop(columns=["Thumbnail", "Notes"], inplace=True)
df.reset_index()
df
```

```python
# Create a new dataframe with all fleet numbers extracted

# Set up new table with no values
bus_df = df[0:0]

# Create df for each set of fleets

for i in df.index:
    if "-" in df.iloc[i,0]:

        fleet_numbers = df.iloc[i,0]
        fleet_list = fleet_numbers.split("-")
        fleet_list = [int(num) for num in fleet_list]
        
        fleet_range = range(fleet_list[0],fleet_list[1]+1)
        fleet_list = list(fleet_range)

        fleet_df = pd.DataFrame({'Fleet number': fleet_list})
    
        fleet_df['Year'] = df.iloc[i,1]
        fleet_df['Manufacturer'] = df.iloc[i,2]
        fleet_df['Model'] = df.iloc[i,3]
        fleet_df['Engine'] = df.iloc[i,4]
        fleet_df['Transmission'] = df.iloc[i,5]
        fleet_df['Destination sign'] = df.iloc[i,6]
        fleet_df['Seating'] = df.iloc[i,7]
        
        bus_df = pd.concat([bus_df, fleet_df])
        
    else:
        
        fleet_number = df.iloc[i,0]
        fleet_number = int(fleet_number)
        fleet_list = [fleet_number]
        
        fleet_df = pd.DataFrame({'Fleet number': fleet_list})
        
        fleet_df['Year'] = df.iloc[i,1]
        fleet_df['Manufacturer'] = df.iloc[i,2]
        fleet_df['Model'] = df.iloc[i,3]
        fleet_df['Engine'] = df.iloc[i,4]
        fleet_df['Transmission'] = df.iloc[i,5]
        fleet_df['Destination sign'] = df.iloc[i,6]
        fleet_df['Seating'] = df.iloc[i,7]
        
        
        bus_df = pd.concat([bus_df, fleet_df])


bus_df = bus_df[['Fleet number','Year','Manufacturer','Model','Engine','Transmission','Destination sign','Seating']]
bus_df.head()
```

```python
# Export as a pickle to save metadata
bus_df.to_pickle('ttc_fleet.pkl')
```
